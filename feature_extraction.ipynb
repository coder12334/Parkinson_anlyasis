```python
def get_in_air(df):
    data = df["Pressure"].asmatrix()
    return (data<600).astype(int).sum()
```

```python
def out_air_surface(df):
    data = df["Pressure"].asmatrix()
    return (data>600).astype(int).sum()
```

```python
def get_no_strokes(df):
    pressure_data=df['Pressure'].as_matrix()
    on_surface = (pressure_data>600).astype(int)
    return ((np.roll(on_surface, 1) - on_surface) != 0).astype(int).sum()
```

```python
def Total_speed(df):
    total_dis = 0
    duration = df["TimeStamp"].asmatrix()[-1]
    coordinate = df["X","Y","Z"].asmatrix()
    for i in range(10,df[0].shape):
        temp=np.linalg.norm(coordinate[i, :] - coordinate[i-10, :])
        total_dis +=temp
    speed = tot_dis/duration
    return speed
        

    
```

```python
def get_features(f,parkinson_target):
    global Header_row
    data_pat=pd.read_csv(f, sep=';', header=None, names=Header_row)
    data_static = data_pat[data_pat["TestID"]==0] #static spiral
    data_dynm = data_pat[data_pat["TestID"]==1] #dynamic spiral
    data_sctp = data_pat[data_pat["TestID"]==2]  #sctp spiral
    
    intial_timestamp = data_pat["TimeStamp"][0]
    data_pat["TimeStamp"] = data_pat["TimeStamp"]  - intial_timestamp #offset
    
    duration_static = data_static["TimeStamp"].as_matrix()[-1] if data_static.shape[0] else 1
    duration_dynamic = data_dynm["TimeStamp"].as_matrix()[-1] if data_dynm.shape[0] else 1
    duration_sctp = data_sctp["TimeStamp"].as_matrix()[-1] if data_sctp.shape[0] else 1
    
    data_point = []
    data_point.append(get_no_strokes(data_static)) if data_static.shape[0] else 0  #strokes for static spirals
    data_point.append(get_no_strokes(data_dynm)) if data_dynm.shape[0] else 0  #dynamic spiral strokes
    
    data_point.append(Total_speed(data_static)) if data_static.shape[0] else 0 #speed for static spiral
    data_point.append(Total_speed(data_dynm)) if data_dynm.shape[0] else 0 #speed for dynamic spiral
    
    magnitude_vec , magnitude_horz_vel_mag , magnitude_vert_vel_mag , time_diff ,Vecl , Horiz_vec , Vert_vec = find_velocity(data_static) if data_static[0] else (0,0,0,0,0,0,0)
    
    #static spiral in magnitude of velocity and  horizontal and vertical velcoity magnitude
    data_point.extend(magnitude_vec , magnitude_horz_vel_mag , magnitude_vert_vel_mag)
    
    magnitude_vec , magnitude_horz_vel_mag , magnitude_vert_vel_mag , time_diff ,Vecl , Horiz_vec , Vert_vec = find_velocity(data_dynm) if data_dynm[0] else (0,0,0,0,0,0,0)
    
    #dynamic spiral magnitude velcoity
    data_point.extend(magnitude_vec,magnitude_horz_vel_mag,magnitude_vert_vel_mag)
    
    magnitude_acc,magnitude_horz_acc_mag,magnitude_vert_acc_mag,time_diff,acc,Horiz_acc,Vert_acc = find_accerlation(data_static) if data_static[0] else (0,0,0,0,0,0)
    
    #static spiral magnitude accelration
    data_point.extend(magnitude_acc,magnitude_horz_acc_mag,magnitude_vert_acc_mag)
    
    magnitude_acc,magnitude_horz_acc_mag,magnitude_vert_acc_mag,time_diff,acc,Horiz_acc,Vert_acc = find_accerlation(data_dynm) if data_dynm[0] else (0,0,0,0,0,0)
    
    
    #dynamic spiral magnitude acc
    
    data_point.extend(magnitude_acc,magnitude_horz_acc_mag,magnitude_vert_acc_mag)
    
    
    magnitude_jerk,magnitude_horz_jerk_mag,magnitude_vert_jerk_mag,time_diff,jerk,Horiz_jerk,Vert_jerk = jerk(data_static) if data_static.shape[0] else (0,0,0,0,0,0)
    
    #jerk hori and vert static spiral
    data_point.extend(magnitude_jerk,magnitude_horz_jerk_mag,magnitude_vert_jerk_mag)
    
    magnitude_jerk,magnitude_horz_jerk_mag,magnitude_vert_jerk_mag,time_diff,jerk,Horiz_jerk,Vert_jerk = jerk(data_dynm) if data_dynm.shape[0] else (0,0,0,0,0,0)
    
    #jerk dynm
    data_point.extend(magnitude_jerk,magnitude_horz_jerk_mag,magnitude_vert_jerk_mag)
    
    
    ncv,ncv_Val=NCV_per_halfcircle(data_static) if data_static.shape[0] else (0,0) # NCV for static test
    data_point.append(ncv_Val)
    ncv,ncv_Val=NCV_per_halfcircle(data_dynm) if data_dynm.shape[0] else (0,0) # NCV for dynamic test
    data_point.append(ncv_Val)
    
    nca,nca_Val=NCA_per_halfcircle(data_static) if data_static.shape[0] else (0,0) # NCA for static test
    data_point.append(nca_Val)
    nca,nca_Val=NCA_per_halfcircle(data_dynm) if data_dynm.shape[0] else (0,0) # NCA for dynamic test
    data_point.append(nca_Val)
    
    
    
    data_point.append(get_in_air_time(data_sctp) if data_sctp.shape[0] else 0) # in air time for STCP
    data_point.append(get_on_surface_time(data_static) if data_static.shape[0] else 0) # on surface time for static test
    data_point.append(get_on_surface_time(data_dynm) if data_dynm.shape[0] else 0) # on surface time for dynamic test
    
    data_point.append(parkinson_target)    # traget. 1 for parkinson. 0 for control.
    
    
    
    return data_point   
    
```
